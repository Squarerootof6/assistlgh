```python
from assistlgh import visualize as av
from matplotlib import pyplot as plt
import numpy as np
from cycler import cycler
av.mplstyle_rc()
up = {
'figure.constrained_layout.h_pad': 0.01,
'figure.constrained_layout.hspace': 0.01,
'figure.constrained_layout.use': True,
'figure.constrained_layout.w_pad': 0.01,
'figure.constrained_layout.wspace': 0.01,
'font.family':['serif'],
'axes.labelpad': 0,
'axes.titlepad': 0,
'axes.labelsize': 'medium',
'xtick.top': True,
'xtick.minor.visible': True,
'xtick.major.pad': 2,
'xtick.labelsize': 'small',
'xtick.direction':'in',
'ytick.right': True,
'ytick.minor.visible': True,
'ytick.labelsize': 'small',
'ytick.major.pad': 2,
'ytick.direction':'in',
'figure.titleweight': 'bold',

'savefig.pad_inches': 0.01,
'savefig.bbox': 'tight',
'grid.linestyle': '--',
'grid.alpha': 0.5,
'legend.labelspacing': 0.1,
'legend.handleheight': 0.5,
'legend.handlelength': 2,
'legend.borderpad': 0.3,
'legend.borderaxespad': 0.4,
'legend.handletextpad': 0.3,
}
plt.rcParams.update(up)
fig,ax = plt.subplots(3,3)
ax = ax.reshape(-1)
x=np.linspace(0,2*np.pi)
fig.suptitle('suptitle')
for i in range(9):
    y=np.sin(x+i)
    ax[i].set_title('title')
    ax[i].plot(x,y,label=i+1,marker='o')
    ax[i].set_xlabel('xlabel')
    ax[i].set_ylabel(r'ylabel$M_{\odot}$')
    ax[i].text(0.1,0.9, 'Al/Mg',weight = 'bold' ,transform = ax[i].transAxes)
    
#plt.tight_layout()
fig.savefig('./test.pdf')
plt.legend()
plt.show()
```
